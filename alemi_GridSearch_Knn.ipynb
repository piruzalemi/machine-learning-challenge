```python
target = df["koi_disposition"]
target_names = ["negative", "positive","zero"]
```

```python
data = df.drop("koi_disposition", axis=1)
feature_names = data.columns
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
# Support vector machine linear classifier
from sklearn.svm import SVC 
model = SVC(kernel='linear') # could be "rbf" - radial Base Function
#model = SVC(kernel='rbf')
```

```python
# Support vector machine linear classifier
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5) # Radial Base Function
#model = SVC(kernel='rbf')
```

```python
# Create the GridSearch estimator along with a parameter object containing the values to adjust
from sklearn.model_selection import GridSearchCV
#param_grid = {'C': [1, 5, 10],
#param_grid = {'C': [1, 5],
#              'gamma': [0.0001, 0.001]}
param_grid ={'n_neighbors':[1,3,7]}
#'gamma': [0.0001, 0.001, 0.01]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
# Fit the model using the grid search estimator. 
# This will take the SVC model and try each combination of parameters
grid.fit(X_train, y_train)
```

```python
# List the best parameters for this dataset
print(grid.best_params_)
```

```python
# List the best score
print(grid.best_score_)
```

```python
# Make predictions with the hypertuned model
predictions = grid.predict(X_test)
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
#                            target_names=["01 Confirmed", "02 False Positive","00 Candidate"]))
```

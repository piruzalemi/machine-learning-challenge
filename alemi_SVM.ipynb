```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv(os.path.join("exoplanet_data.csv"))
df.head()
```

```python
data = df.drop("koi_disposition", axis=1)
feature_names = data.columns
data.head()
```

```python
# Set features. This will also be used as your x values.
data = df[['koi_fpflag_nt', 'koi_period', 
                        'koi_time0bk', 'koi_srad', 'koi_duration', 'koi_depth', 'koi_teq']]
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

target = df["koi_disposition"]
# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(target)
encoded_y = label_encoder.transform(target)
encoded_y
```

```python
for label, original_class in zip(encoded_y, target):
     print('Original Class: ' + str(original_class))
     print('Encoded Label: ' + str(label))
     print('-' * 12)
```

```python
target = encoded_y
target_names = ["Candidate", "Confirmed", "False Positive"]
target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
# Support vector machine linear classifier
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model.fit(X_train, y_train)
```

```python
# Model Accuracy
print('Test Acc: %.3f' % model.score(X_test, y_test))
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
predictions = model.predict(X_test)
print(classification_report(y_test, predictions,
                            target_names=target_names))
```
